### Extract task 3 metrics to add to the slides

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === Load CSV ===
df = pd.read_csv("/ghome/group02/Marina/mcv-c6-2026-team2/Week1/src/task3/results/task3_test.csv")
```

```python

# === Get best post_process per model (max ap50) ===
best_per_model = (
    df.loc[df.groupby("model")["ap50"].idxmax()]
    .sort_values("ap50", ascending=False)
)

# Optional: reset index for clean plotting
best_per_model = best_per_model.reset_index(drop=True)

# === Plot ===
plt.figure(figsize=(10, 6))

bars = plt.bar(
    best_per_model["model"],
    best_per_model["ap50"],
    edgecolor="black"
)

# Add value labels on top of bars (nice for slides)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.01,
        f"{height:.3f}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

# Styling (slide-friendly)
plt.title("Best mAP50 per Model", fontsize=16, fontweight="bold")
plt.xlabel("Model", fontsize=12)
plt.ylabel("mAP50", fontsize=12)
plt.ylim(0, best_per_model["ap50"].max() + 0.1)

plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python

# === Keep best post_process per model ===
best_per_model = (
    df.loc[df.groupby("model")["ap50"].idxmax()]
    .sort_values("ap50", ascending=False)
)

# === Seaborn styling ===
sns.set_theme(
    style="whitegrid",
    context="talk"   # makes it slide-friendly (larger fonts)
)

plt.figure(figsize=(11, 8))

# === Barplot ===
# === Create custom colors ===
highlight_model = "Task2-Adaptative"

colors = [
    "#E76F51" if model == highlight_model else "#4C72B0"
    for model in best_per_model["model"]
]

# === Barplot ===
ax = sns.barplot(
    data=best_per_model,
    x="model",
    y="ap50",
    palette=colors
)

# === Add value labels ===
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.3f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold",
        xytext=(0, 5),
        textcoords="offset points"
    )

# === Labels & styling ===
ax.set_title("Best mAP50 per Model", fontsize=18, weight="bold")
ax.set_xlabel("Model", fontsize=14)
ax.set_ylabel("mAP50", fontsize=14)
ax.set(ylim=(0, 1))

plt.xticks(rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
