```python
PATH = "../data/soccernetball/SN-BAS-2025/england_efl/2019-2020/"
FILE = "Labels-ball.json"

CLASS_PATH = "../data/soccernetball/class.txt"
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def class_distribution(classes: dict):
    classes_legend = {"".join([x[0].upper() for x in label.split(' ')]) : label for label in classes.keys()}

    sns.barplot(classes)
    plt.title("Classes distribution in SoccernetBall Dataset")
    plt.xticks(ticks=list(classes_legend.values()), labels=list(classes_legend.keys()))
    plt.tight_layout()
    plt.show()
```

```python
# Read json
# Count all labels
# Generate histogram
import json
import os
from pathlib import Path
import glob


root = PATH
filename = FILE

with open(CLASS_PATH, "r") as f:
    classes = {line.strip(): 0 for line in f.readlines()}

for folder in glob.glob(os.path.join(root, "*")):
    file = os.path.join(folder, filename)
    with open(file, "r") as f:
        d = json.load(f)
    
    for annotation in d["annotations"]:
        classes[annotation["label"]] += 1

class_distribution(classes)
    
```

```python
with open("../data/soccernetball/train.json", "r") as f:
    info = json.load(f)

furbo = [os.path.basename(element["video"]) for element in info]

with open(CLASS_PATH, "r") as f:
    classes = {line.strip(): 0 for line in f.readlines()}

for folder in glob.glob(os.path.join(root, "*")):
    if os.path.basename(folder) not in furbo:
        continue

    file = os.path.join(folder, filename)
    with open(file, "r") as f:
        d = json.load(f)
    
    for annotation in d["annotations"]:
        classes[annotation["label"]] += 1

class_distribution(classes)
```
